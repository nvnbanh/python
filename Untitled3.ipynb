```python
gpa = float(input("Nhập GPA (0 - 10): "))
if gpa >= 8.5:
    print("Xuất sắc")
elif gpa >= 7.0:
    print("Giỏi")
elif gpa >= 5.5:
    print("Khá")
elif gpa >= 4.0:
    print("Trung bình")
else:
    print("Yếu")
```

```python
n = int(input("Nhập số nguyên dương n: "))
tong = 0
for i in range(1, n + 1):
    if i % 2 == 0:
        tong += i
print("Tổng các số chẵn từ 1 đến", n, "là:", tong)
```

```python
mat_khau_dung = "python123"
so_lan_nhap = 0
while so_lan_nhap < 3:
    mat_khau = input("Nhập mật khẩu: ")
    if mat_khau == mat_khau_dung:
        print("Đăng nhập thành công")
        break
    else:
        so_lan_nhap += 1
if so_lan_nhap == 3:
    print("Tài khoản bị khóa")
```
