```python
n = int(input("Nhập số lượng phần tử n: "))
ds = []
for i in range(n):
    x = float(input(f"Nhập số thứ {i+1}: "))
    ds.append(x)

tong = sum(ds)
tbc = tong / n
print("Trung bình cộng =", tbc)
```

```python
dem_chan = 0
for x in ds:
    if x % 2 == 0:
        dem_chan += 1
print("Số lượng số chẵn =", dem_chan)

```

```python
max_val = min_val = ds[0]
for x in ds:
    if x > max_val:
        max_val = x
    if x < min_val:
        min_val = x
print("Giá trị lớn nhất =", max_val)
print("Giá trị nhỏ nhất =", min_val)
```

```python
print("\nNhập 5 số để sắp xếp:")
a = []
for i in range(5):
    a.append(float(input(f"Số {i+1}: ")))

for i in range(4):
    for j in range(i + 1, 5):
        if a[i] > a[j]:
            a[i], a[j] = a[j], a[i]

print("Dãy sau khi sắp xếp tăng dần:", a)
```

```python
def giai_thua(n):
    if n == 0 or n == 1:
        return 1
    return n * giai_thua(n - 1)

k = int(input("\nNhập số cần tính giai thừa: "))
print(f"{k}! =", giai_thua(k))
```
