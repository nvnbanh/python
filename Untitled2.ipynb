```python
n = int(input())
tong = 0
for i in range(2, n + 1, 2):
    tong += i
print(tong)
```

```python
s = input("Nhap chuoi: ")
words = s.split()
so_tu = len(words)
tu_dai_nhat = words[0]
for tu in words:
    if len(tu) > len(tu_dai_nhat):
        tu_dai_nhat = tu
print("Words:", so_tu)
print("Longest word:", tu_dai_nhat)
```

```python
nums = list(map(int, input("Nhap danh sach: ").split()))
nums = list(set(nums))
nums.sort()
second_largest = nums[-2]
print("Sorted list:", nums)
print("Second largest:", second_largest)
```

```python
scores = list(map(int, input("Nhap diem: ").split()))
valid_scores = []
for s in scores:
    if 0 <= s <= 10:
        valid_scores.append(s)
average = sum(valid_scores) / len(valid_scores)
passed = 0
for s in valid_scores:
    if s >= 5:
        passed += 1
print("Average:", average)
print("Passed:", passed)
```
